```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
datos = pd.read_csv('/content/sample_data/grupo1.csv')
print(datos.head())
print(datos.describe())
plt.hist(datos['Calificacion'], bins=10, color='pink', edgecolor='white')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de las Calificaciones')
plt.show()
```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
# VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='yellow', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='black', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Califs')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables?
El coeficiente de correlación (denotado como
r) mide la fuerza y la dirección de la relación lineal entre dos variables. Sus valores oscilan entre -1 y 1:

Un r de 1 indica una relación lineal perfecta y positiva: a medida que una variable aumenta, la otra también lo hace.
Un r de -1 indica una relación lineal perfecta y negativa: a medida que una variable aumenta, la otra disminuye.
Un r de 0 indica que no hay relación lineal entre las variables.
En el contexto del tiempo de estudio y las calificaciones finales, un coeficiente de correlación positivo sugiere que a mayor tiempo de estudio, mayores son las calificaciones, y un coeficiente negativo indicaría lo contrario.
3. ¿El modelo de regresión lineal es estadísticamente significativo?
ara determinar si el modelo de regresión lineal es estadísticamente significativo, se utiliza generalmente la prueba F para el modelo en su conjunto y la prueba t para los coeficientes individuales.

Prueba F: Evalúa si al menos uno de los coeficientes del modelo es diferente de cero. Si el valor p asociado con la prueba F es menor que el nivel de significancia (típicamente 0.05), se concluye que el modelo es estadísticamente significativo.

Prueba t: Evalúa si un coeficiente individual es diferente de cero. Si el valor p de B1 (el coeficiente del tiempo de estudio) es menor que 0.05, podemos concluir que el tiempo de estudio tiene un efecto significativo en las calificaciones finales.
4. ¿Qué tan confiable es el modelo de regresión para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?
Para evaluar la confiabilidad del modelo de regresión que predice las calificaciones finales de los estudiantes basándose en el tiempo de estudio, considera los siguientes aspectos:

Coeficiente de Determinación (R2)Un R2
  alto (cercano a 1) indica que el modelo explica bien la variabilidad en las calificaciones.

Significancia Estadística de los Coeficientes: Valores p menores que 0.05 sugieren que el tiempo de estudio tiene un efecto significativo en las calificaciones finales.

Error Estándar de la Estimación: Un error estándar pequeño indica predicciones precisas y cercanas a los valores observados.

Análisis de Residuales: Los residuales deben ser normalmente distribuidos, mostrar homoscedasticidad (varianza constante) y no tener patrones sistemáticos.

Validación Cruzada: Ayuda a evaluar el desempeño del modelo en datos no vistos y previene el sobreajuste.

Intervalos de Confianza: Intervalos estrechos para los coeficientes indican precisión en las estimaciones.

En resumen, un modelo es confiable si tiene un R2 alto, coeficientes significativos, un error estándar bajo, residuales apropiados, buen desempeño en validación cruzada y intervalos de confianza estrechos.
