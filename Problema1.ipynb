# Ejercicio 2
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la meia es de 70 kilogramos y la desviacion estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente.
    

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# parametro de la distribucion
Media = 70 #Media
Desviacion_estandar = 4 #Desviacion estandar
TotaldeEstudiantes = 500 #Total de estudiantes

# Variable para calculo de la probabilidad de que un estudiante que pese de 75 kg
Peso_Limite = 79
# Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funcion de distribucion acumulada
#(CDF) de la distribucion normal.
Prob_menos_75kg = norm.cdf(Peso_Limite, Media, Desviacion_estandar)
Estudiantes_mayor_75kg = TotaldeEstudiantes * Prob_menos_75kg
print("Estudiantes que pesan menos de 75 kg:", Prob_menos_75kg)
print("Estudiantes que pesan mas de 75 kg:", Estudiantes_mayor_75kg)
x = np.linspace(Media - 4* Desviacion_estandar, Media + 4* Desviacion_estandar, 1000)
y = norm.pdf(x, Media, Desviacion_estandar)
plt.plot(x, y, 'b', linewidth=2) # Corrected the typo here
plt.fill_between(x, y, where=(x<=Peso_Limite), facecolor='skyblue', alpha=0.5)
plt.title( 'Probabilidad de que un estudiante pese menos de 75 kg. \n  $\mu=70$, $\sigma=4$, x=500   ')
plt.xlabel('Peso (kg)')
plt.ylabel('Dencidad de Probabilidad')
plt.grid(True)
plt.show()
```

#Cuantos pesan menos de 62 kg

```python
plt.figure(figsize=(6, 4))

#parametros de la distribucion normal
Media = 70 #media
Desviacion_estandar= 4 #desviacion estandar
TotaldeEstudiantes = 500 #total de estudiantes

#Variable para el calculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 62

#Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_62kg = norm.cdf(peso_limite, Media, Desviacion_estandar)
#Calculo de el numero de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg = TotaldeEstudiantes * prob_menos_62kg
print("Probabilidad de que un Estudiante pese menos de 62 Kg : ", prob_menos_62kg)
print("Numero de Estudiantes que pesan menos de 75 Kg : ", estudiantes_menos_62kg)


#Calcular la distribucion normal
x = np.linspace(Media - 4* Desviacion_estandar, Media + 4* Desviacion_estandar, TotaldeEstudiantes)
y = norm.pdf(x, Media, Desviacion_estandar)

#Graficar
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x<=peso_limite), facecolor='skyblue', alpha=0.5)
plt.title( 'Probabilidad de que un estudiante pese menos de 62 kg. \n  $\mu=70$, $\sigma=4$, x=500   ')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show
```

#Cuantos pesan más de 79 kg

```python
plt.figure(figsize=(6, 4))

#parametros de la distribucion normal
Media = 70 #media
Desviacion_estandar = 4 #desviacion estandar
Peso_limite = 79 #Peso limite
TotaldeEstudiantes = 500 #total de estudiantes

#Calcular la probabilidad de que un estudiante pese menos o igual de 79 kg
prob_menos_79 = norm.cdf(Peso_Limite, Media, Desviacion_estandar)

#Calcular el numero de estudiantes que pesan mas de 79 kg
prob_mas_79 = 1 - prob_menos_79

#Calculamos el numero aproximado de estudiantes que pesan más de 79 kg
Estudiantes_mas_79 = TotaldeEstudiantes * prob_mas_79
print("Probabilidad de que un Estudiante pese mas de 79 Kg : ", prob_mas_79)
print("Numero de Estudiantes que aproximadamente pesan mas de 79 Kg : ", Estudiantes_mas_79)

#Calcular la funcion de densidad de probabilidad de la distribucion normal
x = np.linspace(Media - 4* Desviacion_estandar, Media + 4* Desviacion_estandar, TotaldeEstudiantes)
y = norm.pdf(x, Media, Desviacion_estandar)

#Graficar la distribucion normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x>=Peso_Limite), facecolor='skyblue', alpha=0.5)
plt.title( 'Probabilidad de que un estudiante pese mas de 79 kg. \n  $\mu=70$, $\sigma=4$, x=500   ')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show
```
